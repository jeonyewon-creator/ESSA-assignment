###실행1_1_3


```python
#라이브러리 별명으로 import
import numpy as np
import scipy as sp
data1 = np.random.normal(0, 1, 100) #난수 생성
summary = sp.stats.describe(data1) #describe 요약통계량 (개수, 평균, 분산, 최소, 최대, 첨도, 왜도)
print(summary)
print(data1)
```

```python
#scipy의 stats 모듈을 import
import numpy as np
from scipy import stats
data1 = np.random.normal(0, 1, 100)
summary = stats.describe(data1) #더 간단하게 사용
print(summary) #난수 생성한 결과로 실행할 때마다 결과 달라짐
```
