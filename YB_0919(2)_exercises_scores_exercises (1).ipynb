````python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
df= pd.DataFrame(raw_data)
print(df)
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
import matplotlib.pyplot as plt
plt.scatter(x='preTestScore', y='postTestScore', data=df, s=df['age'])
plt.title('Relation of preTestScore and postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.scatter(x=df['preTestScore'], y=df['postTestScore'], s=df['postTestScore'] * 4.5, c=df['female'])

plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()


#Seaborn scatterplot() 함수만으로는 sizes 인자에 직접 4.5배를 곱한 값을 사용할 수 없음
#size 인자에 df['postTestScore'] * 4.5처럼 계산된 시리즈를 전달하면, 시본은 이 값을 컬럼 이름으로 인식하려고 시도함. 그러나 df['postTestScore'] * 4.5는 데이터프레임의 컬럼이 아니므로 "시리즈를 float으로 변환할 수 없다"는 오류가 발생함.
````

### BONUS: Create your own question and answer it.

다른 조건은 위와 같지만 이번에는 40세 이상인 경우와 그렇지 않은 경우를 나누어 다른 색으로 표시하여라.

````python
df['age_group'] = df['age'].apply(lambda x: 1 if x >= 40 else 0)
plt.scatter(x=df['preTestScore'], y=df['postTestScore'], s=df['postTestScore'] * 4.5, c=df['age_group'])
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')
plt.show()
````
